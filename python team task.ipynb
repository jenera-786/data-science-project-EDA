```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

 teams choose to bat or field after winning the toss.

```python
toss_decisions = matches_df["toss_decision"].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(toss_decisions, labels=toss_decisions.index,  colors=["skyblue", "lightcoral"], startangle=90)
plt.title("Toss Decision Trends (Bat vs Field)")
plt.show()


plt.figure(figsize=(8, 6))
sns.barplot(x=toss_decisions.index, y=toss_decisions.values, palette=["skyblue", "lightcoral"])


plt.xlabel("Toss Decision")
plt.ylabel("Count")
plt.title("Toss Decision Trends (Bat vs Field)")

# Show plot
plt.show()

```

Count matches won by each team per season and visualize trends over years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
season_wins = matches_df.groupby(["Season", "winner"]).size().reset_index(name="wins")
plt.figure(figsize=(12, 6))
sns.lineplot(data=season_wins, x="Season", y="wins", hue="winner", marker="o", palette="tab10")
plt.xlabel("Season")
plt.ylabel("Matches Won")
plt.title("Matches Won by Each Team per Season")
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
plt.figure(figsize=(12, 6))
sns.barplot(data=season_wins, x="Season", y="wins", hue="winner", palette="tab10")
plt.xlabel("Season")
plt.ylabel("Matches Won")
plt.title("Matches Won by Each Team per Season")
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc="upper left")

# Show plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(matches_df["win_by_runs"], bins=20, color="blue", label="Win by Runs")
sns.histplot(matches_df["win_by_wickets"], bins=10, kde=True, color="green", label="Win by Wickets")
plt.xlabel("Win Margin")
plt.ylabel("Frequency")
plt.title("Distribution of Win Margins (Runs & Wickets)")
plt.legend()
plt.show()
plt.figure(figsize=(12, 6))
df_melted = matches_df.melt(value_vars=["win_by_runs", "win_by_wickets"], var_name="Win Type", value_name="Margin")
sns.boxplot(x="Win Type", y="Margin", data=df_melted, palette=["blue", "green"])
plt.xlabel("Win Type")
plt.ylabel("Win Margin")
plt.title("Win Margin Distribution (Runs & Wickets)")
plt.show()
```
